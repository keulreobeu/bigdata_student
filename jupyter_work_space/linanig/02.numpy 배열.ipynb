```python
# 가중평균: 두 백터가 닮음정도를 정략적으로 나타낸값 비슷할수록 유사도가 커짐
digits = load_digits()
d1 = digits.images[0]
d2 = digits.images[10]
d3 = digits.images[1]
d4 = digits.images[11]
v1 = d1.reshape(64,1)
v2 = d2.reshape(64,1)
v3 = d3.reshape(64,1)
v4 = d4.reshape(64,1)
```

```python
# gridspec은 피겨의 비율을 상세하게 조절하기위해 사용.
plt.figure(figsize=(9,9))
gs = gridspec.GridSpec(1,8,height_ratios=[1],
                      width_ratios=[9,1,9,1,9,1,9,1])

for i in range(4):
    plt.subplot(gs[2*i])
    plt.imshow(eval('d'+str(i+1)), aspect=1, cmap=plt.cm.bone_r)
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    plt.title(f'image {i + 1}')
    plt.subplot(gs[2*i+1])
    plt.imshow(eval('v'+str(i+1)), aspect=0.25, cmap=plt.cm.bone_r)
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    plt.title(f'vector {i + 1}')
    
    
plt.tight_layout()
plt.show()
```

```python
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])
C = np.array([[9,10],[11,12]])
print('교환법칙 X')
print(A@B)
print(B@A)
print('분배법칙 o')
print(A@(B+C))
print(A@B + A@C)
print('분배법칙 o')
print((A+B)@C)
print(A@C + B@C)
```

```python
print('곱의 역행렬은 연산 순서가 바뀐다.')
print((A@B).T)
print(B.T@A.T)
```
