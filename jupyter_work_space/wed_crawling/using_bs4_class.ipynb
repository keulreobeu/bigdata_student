```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://news.daum.net/'
response = requests.get(url)
bs = BeautifulSoup(response.text, 'html.parser')
for tag in bs.find_all(class_='item_issue'):
    print(tag)
```

```python
#공백문자가 계속 들어가서 seq % 2 를 사용함.
it_url = 'https://news.daum.net/digital#1'
it_response = requests.get(it_url)
it_bs = BeautifulSoup(it_response.text, 'html.parser')

for tag in it_bs.find_all(class_='list_newsmajor'):
    for seq, li in enumerate(tag.children):
        if seq % 2 == 1:
            print(list(li.children)[1].text.replace('\n', ' '))
```

```python
url = 'https://news.daum.net/digital#1'
response = requests.get(url)
bs = BeautifulSoup(response.text, 'html.parser')
result = {}
for ul in bs.find_all(class_='list_newsmajor'):
    for strong in ul.find_all(class_='tit_g'):
        print(f'{strong.a.get("href")}: {strong.a.text}')
        result[strong.a.get("href")] = strong.a.text
print(result)
```

```python
import json

with open('daum_it_news.json','w', encoding='utf=8') as f:
    json.dump(result, f)
```

```python
with open('daum_it_news.json', 'r', encoding='utf=8') as f:
    contents = f.read()
    json_data = json.loads(contents)
    print(json_data)
```
