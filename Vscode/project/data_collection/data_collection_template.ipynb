## [의약품 사용 통계]

### 1.   데이터 정의

'보건의료빅데이터개방시스템'에서 제공하는 의약품 사용량을 월별로 집계한 통계입니다.

### 2. 수집 방법





'보건의료빅데이터개방시스템'에서 제공하는 xls파일을 받아옵니다.



### 3.   수집 경로

'보건의료빅데이터개방시스템'의 의약품 사용 통계

약국 직접/처방 조제량   약효분류별 데이터를 받는다.

http://opendata.hira.or.kr/op/opc/olapMsupInfo.do

### 4.   데이터의 논리적 구조

처방조제수 월별기간을 index로 환자수, 수량, 사용금액을 column으로 연속형 데이터 구조이다.

약효분류별 월별기간을 index로 '해열,진통,소염제', '진해거담제'을 column으로 연속형 데이터 구조이다.

### 5.   데이터의 물리적 저장 구조

xls방식의 데이터
